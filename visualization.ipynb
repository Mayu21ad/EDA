```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r'D:\Visualizations\IRIS\IRIS.csv')
```

```python
data.head()
```

```python
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(10,10))
plt.plot(data['sepal_length'], data['petal_length'], 'go', markersize=10)
plt.plot(data['sepal_width'], data['petal_width'], 'bv', markersize=10)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.xticks(np.arange(1,10,0.5))
plt.yticks(np.arange(1,10,0.5))

plt.plot(data['sepal_length'], data['petal_length'], 'go', alpha=0.5, markersize=10)
plt.plot(data['sepal_width'], data['petal_width'], 'bv', alpha=0.2, markersize=10)
plt.show()
```

```python
plt.plot(data['sepal_length'], data['petal_length'], 'b-', alpha=0.5)
plt.show()
```

```python
plt.plot(data['sepal_length'], data['petal_length'], 'b:', alpha=0.5)
plt.show()
```

```python
plt.plot(data['sepal_width'], data['petal_width'], 'b', 'o', alpha=0.5, markersize=1, linewidth=15)
plt.show()
```

```python
plt.plot(data['sepal_length'], data['petal_length'], 'r', '-', linewidth=15, alpha=0.5, markersize=10)
plt.plot(data['sepal_length'], data['petal_length'], 'g', 'o')
```
